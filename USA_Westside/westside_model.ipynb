```python
# Core tables
events = pd.read_csv(base / "events.csv")[["event_id", "duration"]].copy()
events["duration"] = events["duration"].astype(int)

constraints = pd.read_csv(base / "constraints.csv")
constraint_params = pd.read_csv(base / "constraint_params.csv")
constraint_applied = pd.read_csv(base / "constraint_applies_to.csv")

# 1) Keep only split constraints
split_cons = constraints.loc[
    constraints["constraint_type"].eq("SplitEventsConstraint"),
    ["constraint_id", "required", "weight", "cost_function"]
].copy()

# 2) Extract split parameters (min/max duration + amount)
needed_paths = {
    "MinimumDuration": "min_duration",
    "MaximumDuration": "max_duration",
    "MinimumAmount": "min_amount",
    "MaximumAmount": "max_amount",
}

split_params = (
    constraint_params[constraint_params["constraint_id"].isin(split_cons["constraint_id"])]
    .loc[lambda d: d["path"].isin(needed_paths.keys()), ["constraint_id", "path", "value"]]
    .assign(path=lambda d: d["path"].map(needed_paths), value=lambda d: d["value"].astype(int))
    .pivot_table(index="constraint_id", columns="path", values="value", aggfunc="first")
    .reset_index()
)

split_specs = split_cons.merge(split_params, on="constraint_id", how="left")

# 3) Which events each split constraint applies to
split_applied = (
    constraint_applied[
        (constraint_applied["constraint_id"].isin(split_specs["constraint_id"]))
        & (constraint_applied["path"] == "AppliesTo/Events/Event")
    ][["constraint_id", "reference"]]
    .rename(columns={"reference": "event_id"})
)

# 4) Event + split spec rows
event_split = (
    split_applied
    .merge(events, on="event_id", how="left")
    .merge(split_specs, on="constraint_id", how="left")
)

# Guard: each event should be controlled by at most one split constraint
dup_events = event_split["event_id"][event_split["event_id"].duplicated()].unique().tolist()
if dup_events:
    raise ValueError(f"Events with multiple split constraints: {dup_events[:10]} ...")

def choose_split_durations(total_d, min_amt, max_amt, min_dur, max_dur):
    """
    Find integer sub-event durations s.t.
      n in [min_amt, max_amt]
      each dur in [min_dur, max_dur]
      sum(durs) = total_d
    Returns list of durations or None if impossible.
    """
    for n in range(min_amt, max_amt + 1):
        if n * min_dur <= total_d <= n * max_dur:
            base = total_d // n
            rem = total_d % n
            durs = [base + 1] * rem + [base] * (n - rem)
            if all(min_dur <= x <= max_dur for x in durs):
                return durs
    return None

rows = []
errors = []

# 5) Build sub-events for constrained events
for r in event_split.itertuples(index=False):
    durs = choose_split_durations(
        total_d=int(r.duration),
        min_amt=int(r.min_amount),
        max_amt=int(r.max_amount),
        min_dur=int(r.min_duration),
        max_dur=int(r.max_duration),
    )

    if durs is None:
        if str(r.required).upper() == "TRUE":
            errors.append(
                {
                    "event_id": r.event_id,
                    "constraint_id": r.constraint_id,
                    "duration": int(r.duration),
                    "min_amount": int(r.min_amount),
                    "max_amount": int(r.max_amount),
                    "min_duration": int(r.min_duration),
                    "max_duration": int(r.max_duration),
                }
            )
            continue
        else:
            durs = [int(r.duration)]  # soft violation fallback

    for k, sd in enumerate(durs, start=1):
        rows.append(
            {
                "parent_event_id": r.event_id,
                "sub_event_id": f"{r.event_id}__sub{k}",
                "sub_index": k,
                "sub_duration": int(sd),
                "split_constraint_id": r.constraint_id,
                "required": r.required,
                "weight": r.weight,
                "cost_function": r.cost_function,
                "was_split": len(durs) > 1,
            }
        )

# 6) Add unconstrained events as single sub-events
constrained_events = set(event_split["event_id"].tolist())
for r in events.itertuples(index=False):
    if r.event_id in constrained_events:
        continue
    rows.append(
        {
            "parent_event_id": r.event_id,
            "sub_event_id": f"{r.event_id}__sub1",
            "sub_index": 1,
            "sub_duration": int(r.duration),
            "split_constraint_id": None,
            "required": None,
            "weight": None,
            "cost_function": None,
            "was_split": False,
        }
    )

sub_events = pd.DataFrame(rows).sort_values(["parent_event_id", "sub_index"]).reset_index(drop=True)

if errors:
    df_split_errors = pd.DataFrame(errors)
    print("Required split constraints with no feasible split found:")
    print(df_split_errors.head(20))
else:
    df_split_errors = pd.DataFrame()
    print("All required split constraints produced feasible sub-events.")

print(f"Parents: {events['event_id'].nunique()}, Sub-events: {sub_events['sub_event_id'].nunique()}")
```

```python
### Feasible Start times for events:

#T_index = {t:i for i,t in enumerate(T)}

time_to_day = dict(zip(df_times["time_id"], df_times["day_ref"]))

# Build feasible starts per event (no crossing day boundary) (implicitely covers constraints 5,9,10,11,12)
feasible_starts = {}
for se in SE:
    d = int(subevent_to_duration[se]) # number of periods needed
    starts = []
    for ts in T: # for all possible start times
        i0 = T_index[ts]
        # must have enough periods remaining in the global list
        if i0 + d - 1 >= len(T):
            continue
        day0 = time_to_day[ts]
        block = T[i0:i0 + d]
        # enforce all times in the block are same day
        if all(time_to_day[t] == day0 for t in block):
            starts.append(ts)
    feasible_starts[se] = starts

# If event e starts at ts, then occ_times[(e, ts)] = [t1, t2, ...] is the list of all times occupied by that event
occ_times = {}
for se in SE:
    d = int(subevent_to_duration[se])
    for ts in feasible_starts[se]:
        i0 = T_index[ts]
        occ_times[(se, ts)] = T[i0:i0 + d]


# Helper: all (e, ts) pairs that occupy a given time t
# ie for Fr_3, if event 1 is duration 1 and event 2 is duration 2, then (1, Fr_3) and (2, Fr_2) and (2, Fr_3) occupy it
occupies_at_time = {t: [] for t in T}
for se in SE:
    for ts in feasible_starts[se]:
        for t in occ_times[(se, ts)]:
            occupies_at_time[t].append((se, ts))

```

### 2b: Pyomo Model part 1: Sets and Decision Variables

Constraints:

Part 1: Event -> Time (hard constraints only)
1) AssignTimes: [gr_EventsGeneral] Every event must be assigned a time
2) NoResourceClashes: No clashes for any of the resources
3) SplitEventsCOnstraint_number: Events from selected event group should be split between a min and max number of subevents, with each subevent being between a min and max number of time periods
4) SameDay_Duration(2/3/4): Events of duration 2,3 and 4 must start at valid times (no bleeding into the next day)

Extra: Number of events at any one time should be less than number of rooms available

Part 2: Event -> Time (with soft constraints)
1) PreferTimesConstraint(waffle): sub-events of selected event and of given duration have preffered times in which they are scheduled (start times) (soft: weight = 4, linear)
2) SpreadEventsConstraint_course: between a min and max (day group specific) number of sub-events of selected event can occur on a given day (soft: weight=1, linear)

Part 3: Event -> Room 
1) Every event must be assigned exactly 1 room
2) At most one event can occupy a room at a time
3) PreferResourcesConstraint_Room: Selected event prefers to be scheduled with selected room or room (soft: weight=1, linear)
4) PreferResourcesConstraint_Dept: Selected event prefers to be scheduled with selected room group room (soft: weight=1, linear)


### 2b: Model 1

```python
# ---- Model ----
m1 = pyo.ConcreteModel()

# ---- Sets ----
m1.E = pyo.Set(initialize=E) # Set of events
m1.T = pyo.Set(initialize=T, ordered=True) # Set of timeslots (ordered)
m1.R = pyo.Set(initialize=R) # Set of rooms

m1.Teachers = pyo.Set(initialize=all_teachers) # set of all teachers
m1.Classes = pyo.Set(initialize=all_classes) # set of all classes
m1.Students = pyo.Set(initialize=all_students)

m1.SE = pyo.Set(initialize=SE) # Set of events
# standard set up
m1.X_index = pyo.Set(
    dimen=2,
    initialize=[(se, ts) for se in SE for ts in feasible_starts[se]]
)



# ---- Decision Variables ----
# Start-time decision variables:
# x[e, ts, r] = 1 if event e starts at time ts in room r
# Only define variables for feasible starts (and all rooms) (THIS SPEEDS UP MODEL)
m1.x = pyo.Var(m1.X_index, domain=pyo.Binary)
```

```python

# Standard set up

# 1) Each event scheduled exactly once (one start time)   (covers constraints 4)
def one_start_rule(m, se):
    return sum(m.x[se, ts] for ts in feasible_starts[se]) == 1
m1.OneStart = pyo.Constraint(m1.SE, rule=one_start_rule)


# 3) No teacher conflicts: a teacher can teach at most one event at a time
def teacher_conflict_rule(m, teacher, t):
    terms = [
        m.x[se, ts]
        for (se, ts) in occupies_at_time.get(t, [])
        if teacher in as_list(subevent_to_teachers[se])
    ]
    if not terms:
        return pyo.Constraint.Skip   # or pyo.Constraint.Feasible
    return sum(terms) <= 1

m1.TeacherConflict = pyo.Constraint(all_teachers, m1.T, rule=teacher_conflict_rule)


#m1.ClassConflict = pyo.Constraint(all_classes, m1.T, rule=class_conflict_rule)


#m1.StudentConflict = pyo.Constraint(all_students, m1.T, rule=student_conflict_rule)


# 5) Max number of simultaneous events cannot exceed number of rooms
m1.N_rooms = pyo.Param(initialize=len(R), within=pyo.NonNegativeIntegers)

def total_events_vs_rooms_rule(m, t):
    terms = [m.x[se, ts] for (se, ts) in occupies_at_time.get(t, [])]
    if not terms:
        return pyo.Constraint.Skip
    return sum(terms) <= m.N_rooms

m1.TotalEventsVsRooms = pyo.Constraint(m1.T, rule=total_events_vs_rooms_rule)

```

### Solve

```python
# ---- Objective ---- (hard constraints only)
m1.obj = pyo.Objective(expr=0, sense=pyo.minimize)
```

```python
seeds = [10, 20, 30, 40, 50]

solve_times = []
objectives = []
num_vars = []
num_cons = []
solve_status = []
start_x_list = []

import time as time_module

for seed in seeds:
    solver = pyo.SolverFactory("gurobi")
    solver.options["TimeLimit"] = 600
    solver.options["Seed"] = seed
    solver.options["Threads"] = 1

    # Manual timing
    start_time = time_module.time()
    res = solver.solve(m1, tee=False)
    elapsed_time = time_module.time() - start_time

    solve_times.append(elapsed_time)
    objectives.append(pyo.value(m1.obj))
    num_vars.append(m1.nvariables())
    num_cons.append(m1.nconstraints())
    solve_status.append(res.solver.termination_condition)

    # Store solution
    seed_solution = {}
    for idx in m1.X_index:
        val = pyo.value(m1.x[idx], exception=False)
        seed_solution[idx] = 0.0 if val is None else float(val)
    start_x_list.append(seed_solution)

# Keep compatibility with later cells that expect start_x
start_x = start_x_list[0]

# Calculate averages
avg_time = sum(solve_times) / len(solve_times)
avg_obj = sum(objectives) / len(objectives)
avg_vars = sum(num_vars) / len(num_vars)
avg_cons = sum(num_cons) / len(num_cons)

print("=" * 50)
print("SOLVER METRICS SUMMARY - PHASE 1")
print("=" * 50)
print(f"Average Solve Time (s): {avg_time:.4f}")
print(f"Average Objective: {avg_obj:.6f}")
print(f"Average Variables: {avg_vars:.0f}")
print(f"Average Constraints: {avg_cons:.0f}")
print(f"Termination Status: {set(solve_status)}")
print("=" * 50)

print(f"\nPhase 1 produced {len(start_x_list)} solutions (one per seed)")
for i, seed_sol in enumerate(start_x_list):
    scheduled = sum(1 for v in seed_sol.values() if v > 0)
    print(f"  Seed {seeds[i]}: {scheduled} events scheduled in {solve_times[i]:.2f}s, Obj={objectives[i]:.6f}")
```

```python
def add_prefer_times_penalty(model, x):
    """
    Assumes these are already defined in outer scope:
      cids, cid_to_weight, pref_times, dur_filter,
      applied_events, parent_to_sub, sub_duration, feasible_starts
    """
    penalty = 0.0

    for cid in cids:
        w = 4
        pref = pref_times.get(cid, set())

        for parent_event in applied_events.get(cid, set()):
            for se in parent_to_sub.get(parent_event, []):
                d_se = int(sub_duration[se])
                starts = feasible_starts.get(se, [])
                if not starts:
                    continue

                bad_start = sum((0 if ts in pref else 1) * x[se, ts] for ts in starts)
                dev_se = d_se * bad_start
                penalty += w * dev_se

    model.PreferredTimesPenalty = pyo.Expression(expr=penalty)
    return model.PreferredTimesPenalty

    

```

```python
# 1) pick spread constraints
cdf = df_constraints[df_constraints["constraint_type"] == "SpreadEventsConstraint"].copy()
cdf = cdf[cdf["required"].astype(str).str.upper() != "TRUE"] # dataframe of spreadevents constraints

spread_cids = set(cdf["constraint_id"]) #constraint IDs
cid_to_weight = cdf.set_index("constraint_id")["weight"].astype(float).to_dict() # weight of each constraint

# 2) time_group -> set(times)
tg_to_times = (
    df_time_group_membership.groupby("group_id")["time_id"].apply(set).to_dict()
)

# 3) parse per-constraint TimeGroups/TimeGroup + Minimum + Maximum
    #    (flattened CSV needs ordered scan)
cp = df_constraint_params[df_constraint_params["constraint_id"].isin(spread_cids)].copy()
cp = cp.reset_index(drop=False).rename(columns={"index": "_row_order"})
cp = cp.sort_values(["constraint_id", "_row_order"])

cid_to_tg_bounds = {}  # cid -> list[(time_group_id, min, max)]
for cid, g in cp.groupby("constraint_id", sort=False):
    rows = g[["path", "value"]].itertuples(index=False)
    cur_tg = None
    cur_min = None
    cur_max = None
    out = []
    for r in rows:
        if r.path == "TimeGroups/TimeGroup":
            if cur_tg is not None:
                out.append((cur_tg, int(cur_min), int(cur_max)))
            cur_tg = r.value
            cur_min = 0
            cur_max = 10**9
        elif r.path == "TimeGroups/TimeGroup/Minimum":
            cur_min = int(r.value)
        elif r.path == "TimeGroups/TimeGroup/Maximum":
            cur_max = int(r.value)
    if cur_tg is not None:
        out.append((cur_tg, int(cur_min), int(cur_max)))
    cid_to_tg_bounds[cid] = out


```

```python
# 4) map each constraint -> each event 
applies_eg = df_constraint_applied[
    (df_constraint_applied["constraint_id"].isin(spread_cids)) &
    (df_constraint_applied["path"] == "AppliesTo/EventGroups/EventGroup")
][["constraint_id", "reference"]]

# map constraint -> events (list)
cid_to_eventgroups = applies_eg.groupby("constraint_id")["reference"].apply(list).to_dict()

# 5) event_group -> parent events, parent event -> sub-events
eg_to_events = (
    df_event_groups_membership.groupby("event_group_id")["event_id"].apply(set).to_dict()
)
parent_to_sub = (
    df_sub_events.groupby("parent_event_id")["sub_event_id"].apply(list).to_dict()
)

# 6) build index triples (cid, event_group, time_group)
triples = []
for cid in spread_cids: # for each constraint
    for eg in cid_to_eventgroups.get(cid, []): # for event group in constraint
        for tg, mn, mx in cid_to_tg_bounds.get(cid, []): # for time, max, min 
            triples.append((cid, eg, tg, int(mn), int(mx)))

#if not triples:
    #model.spread_events_penalty = pyo.Expression(expr=0.0)
    #return model.spread_events_penalty

#applies_eg
#cid_to_eventgroups
#eg_to_events
#parent_to_sub
#triples
```

```python
# Stable keys for Pyomo sets/vars
keys = [(cid, eg, tg) for cid, eg, tg, _, _ in triples]
bounds_map = {(cid, eg, tg): (mn, mx) for cid, eg, tg, mn, mx in triples}

m1.SpreadIdx = pyo.Set(initialize=keys, dimen=3)
m1.spread_under = pyo.Var(m1.SpreadIdx, domain=pyo.NonNegativeReals)
m1.spread_over = pyo.Var(m1.SpreadIdx, domain=pyo.NonNegativeReals)

def count_expr(cid, eg, tg):
    tg_times = tg_to_times.get(tg, set())
    parent_events = eg_to_events.get(eg, set())
    expr = 0.0
    for pe in parent_events: # for each parent event in constraint group
        for se in parent_to_sub.get(pe, []): #for each sub event of parent event
            for ts in feasible_starts.get(se, []): #for all feasible starts
                if ts in tg_times:
                    expr += m1.x[se, ts]
    return expr

    # under >= min - count
def under_rule(m, cid, eg, tg):
    mn, mx = bounds_map[(cid, eg, tg)]
    return m.spread_under[cid, eg, tg] >= mn - count_expr(cid, eg, tg)
m1.SpreadUnderConstr = pyo.Constraint(m1.SpreadIdx, rule=under_rule)

    # over >= count - max
def over_rule(m, cid, eg, tg):
    mn, mx = bounds_map[(cid, eg, tg)]
    return m.spread_over[cid, eg, tg] >= count_expr(cid, eg, tg) - mx
m1.SpreadOverConstr = pyo.Constraint(m1.SpreadIdx, rule=over_rule)


# weighted linear penalty
m1.SpreadEventsPenalty = pyo.Expression(
    expr=sum(
        float(cid_to_weight.get(cid, 1.0)) *
        (m1.spread_under[cid, eg, tg] + m1.spread_over[cid, eg, tg])
        for (cid, eg, tg) in m1.SpreadIdx
    )
)
```

```python
import os
import re
import time as time_module
import numpy as np
import matplotlib.pyplot as plt
import pyomo.environ as pyo

TIME_LIMIT = 600
LOG_DIR = "gurobi_phase2_logs"
SENSE = "min"
os.makedirs(LOG_DIR, exist_ok=True)


def clear_all_var_values(model):
    for v in model.component_data_objects(pyo.Var, active=True):
        v.set_value(None, skip_validation=True)


def load_x_warm_start(model, seed_solution, tol=0.5):
    # Start from a clean model state so no stale values are exported
    clear_all_var_values(model)

    assigned = 0
    for idx in model.X_index:
        val = seed_solution.get(idx, 0.0)
        start_val = 1 if val > tol else 0
        model.x[idx].set_value(start_val, skip_validation=True)
        assigned += 1

    return assigned


def parse_model_stats(logfile):
    rows = cols = nnz = None
    pat = re.compile(
        r"Optimize a model with\s+(\d+)\s+rows,\s+(\d+)\s+columns and\s+(\d+)\s+nonzeros"
    )
    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            m = pat.search(line)
            if m:
                rows = int(m.group(1))
                cols = int(m.group(2))
                nnz = int(m.group(3))
                break
    return rows, cols, nnz


def parse_heuristic_trajectory(logfile):
    times = []
    objs = []

    elapsed_pat = re.compile(
        r"Elapsed time for NoRel heuristic:\s*([0-9]+(?:\.[0-9]+)?)s"
    )
    found_pat = re.compile(
        r"Found heuristic solution:\s*objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )
    mip_start_pat = re.compile(
        r"Loaded user MIP start with objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )

    current_time = None

    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()

            m = elapsed_pat.search(s)
            if m:
                current_time = float(m.group(1))
                continue

            m = mip_start_pat.search(s)
            if m:
                times.append(0.0)
                objs.append(float(m.group(1)))
                continue

            m = found_pat.search(s)
            if m:
                times.append(current_time)
                objs.append(float(m.group(1)))
                continue

            if s.startswith("H") or s.startswith("*"):
                toks = s.split()
                if toks[0] == "H" and len(toks) >= 8:
                    objs.append(float(toks[3]))
                    times.append(float(toks[-1].rstrip("s")))
                elif toks[0] == "*" and len(toks) >= 9:
                    objs.append(float(toks[4]))
                    times.append(float(toks[-1].rstrip("s")))

    return times, objs



def parse_first_feasible_time(logfile):
    elapsed_pat = re.compile(
        r"Elapsed time for NoRel heuristic:\s*([0-9]+(?:\.[0-9]+)?)s"
    )
    found_pat = re.compile(
        r"Found heuristic solution:\s*objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )
    mip_start_pat = re.compile(
        r"Loaded user MIP start with objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )

    current_time = None

    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()

            m = elapsed_pat.search(s)
            if m:
                current_time = float(m.group(1))
                continue

            m = mip_start_pat.search(s)
            if m:
                return 0.0, float(m.group(1)), "warm_start"

            m = found_pat.search(s)
            if m:
                return current_time, float(m.group(1)), "heuristic_log"

    return None, None, None


def parse_mip_start_status(logfile):
    accepted = False
    messages = []

    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()
            if "User MIP start" in s or "Loaded user MIP start" in s:
                messages.append(s)
            if "Loaded user MIP start" in s or "User MIP start produced solution" in s:
                accepted = True

    return accepted, messages


phase2_results = []
trajectories = []

for i, seed_solution in enumerate(start_x_list):
    seed = seeds[i]

    print(f"\n{'='*60}")
    print(f"PHASE 2 - Warm Start from Seed {seed}")
    print(f"{'='*60}")

    assigned = load_x_warm_start(m1, seed_solution)

    logfile = os.path.join(LOG_DIR, f"phase2_seed_{seed}.log")
    if os.path.exists(logfile):
        os.remove(logfile)

    solver2 = pyo.SolverFactory("gurobi")
    solver2.options["TimeLimit"] = TIME_LIMIT
    solver2.options["Seed"] = seed
    solver2.options["MIPFocus"] = 1
    solver2.options["Heuristics"] = 0.5
    solver2.options["NoRelHeurTime"] = TIME_LIMIT
    solver2.options["DisplayInterval"] = 1
    solver2.options["LogFile"] = logfile
    solver2.options["Threads"] = 1

    start_time = time_module.time()
    res2 = solver2.solve(m1, tee=False, warmstart=True, load_solutions=False)
    elapsed_time = time_module.time() - start_time

    solver_status = str(res2.solver.status)
    termination = str(res2.solver.termination_condition)

    has_solution = len(res2.solution) > 0
    if has_solution:
        m1.solutions.load_from(res2)

    try:
        obj2 = pyo.value(m1.obj) if has_solution else float("nan")
    except:
        obj2 = float("nan")
        has_solution = False
        

    chosen_start = {}
    if has_solution:
        starts_by_subevent = {}
        for (se, ts) in m1.X_index:
            starts_by_subevent.setdefault(se, []).append(ts)

        for se in m1.SE:
            picked = [
                ts for ts in starts_by_subevent.get(se, [])
                if pyo.value(m1.x[se, ts], exception=False) is not None
                and pyo.value(m1.x[se, ts], exception=False) > 0.5
            ]
            if len(picked) != 1:
                raise ValueError(f"Sub-event {se}: expected exactly 1 chosen start, got {picked}")
            chosen_start[se] = picked[0]




    num_vars = m1.nvariables()
    num_cons = m1.nconstraints()
    rows_log, cols_log, nnz_log = parse_model_stats(logfile)
    tvals, ovals = parse_heuristic_trajectory(logfile)
    first_feas_time, first_feas_obj, first_feas_source = parse_first_feasible_time(logfile)
    mip_start_accepted, mip_start_messages = parse_mip_start_status(logfile)

    if has_solution:
        if not tvals:
            tvals.append(elapsed_time)
            ovals.append(obj2)
        elif abs(ovals[-1] - obj2) > 1e-9 and elapsed_time > (tvals[-1] if tvals[-1] is not None else -1):
            tvals.append(elapsed_time)
            ovals.append(obj2)

    phase2_results.append({
        "seed": seed,
        "solver_status": solver_status,
        "termination": termination,
        "objective": obj2,
        "solve_time": elapsed_time,
        "num_vars": num_vars if num_vars is not None else cols_log,
        "num_cons": num_cons if num_cons is not None else rows_log,
        "num_nonzeros": nnz_log,
        "has_solution": has_solution,
        "assigned_start_values": assigned,
        "chosen_start": chosen_start,
        "mip_start_accepted": mip_start_accepted,
        "mip_start_messages": mip_start_messages,
        "n_heuristic_updates": len(tvals),
        "first_feasible_time": first_feas_time,
        "first_feasible_objective": first_feas_obj,
        "first_feasible_source": first_feas_source,
        "logfile": logfile,
    })


    trajectories.append({
        "seed": seed,
        "times": tvals,
        "objectives": ovals,
    })

    print(f"Assigned x warm-start values: {assigned}")
    print(f"Solver status: {solver_status}")
    print(f"Termination: {termination}")
    print(f"Warm start accepted: {mip_start_accepted}")
    if mip_start_messages:
        print("MIP start log:")
        for msg in mip_start_messages:
            print("  ", msg)
    print(f"Objective: {obj2:.6f}" if has_solution else "Objective: No solution loaded")
    print(f"Solve time: {elapsed_time:.2f}s")
    print(f"Heuristic updates found: {len(tvals)}")
    print(f"Variables: {num_vars}")
    print(f"Constraints: {num_cons}")
    print(f"Nonzeros: {nnz_log}")


# SUMMARY
valid_objs = [r["objective"] for r in phase2_results if not np.isnan(r["objective"])]
avg_time = np.mean([r["solve_time"] for r in phase2_results])

print(f"\n{'='*60}")
print("PHASE 2 SUMMARY")
print(f"{'='*60}")
for r in phase2_results:
    obj_str = f"{r['objective']:.6f}" if not np.isnan(r["objective"]) else "No solution"
    first_time_str = (
        f"{r['first_feasible_time']:.2f}s"
        if r["first_feasible_time"] is not None
        else "Not found"
    )
    first_obj_str = (
        f"{r['first_feasible_objective']:.6f}"
        if r["first_feasible_objective"] is not None
        else "Not found"
    )
    print(
        f"Seed {r['seed']}: "
        f"Status={r['solver_status']}, "
        f"Termination={r['termination']}, "
        f"Obj={obj_str}, "
        f"Time={r['solve_time']:.2f}s, "
        f"FirstFeasibleTime={first_time_str}, "
        f"FirstFeasibleObj={first_obj_str}, "
        f"FirstFeasibleSource={r['first_feasible_source']}, "
        f"HeuristicUpdates={r['n_heuristic_updates']}"
    )

print(f"\nAverage Solve Time (s): {avg_time:.4f}")
print(f"Average Objective: {np.mean(valid_objs):.6f}" if valid_objs else "Average Objective: Not available")

feasible_results = [r for r in phase2_results if not np.isnan(r["objective"])]
if feasible_results:
    best_result = min(feasible_results, key=lambda x: x["objective"]) if SENSE == "min" else max(feasible_results, key=lambda x: x["objective"])
    print(
        f"Best Phase 2 Solution: Seed {best_result['seed']} "
        f"with objective {best_result['objective']:.6f}"
    )
else:
    print("Best Phase 2 Solution: No feasible solution was loaded")


# PLOT
fig, ax = plt.subplots(figsize=(10, 6))

plotted_any = False
for tr in trajectories:
    plot_times = [0.0 if t is None else t for t in tr["times"]]
    plot_pairs = sorted(zip(plot_times, tr["objectives"]), key=lambda x: x[0])

    if plot_pairs:
        plot_times, plot_objs = zip(*plot_pairs)

        ax.step(
            plot_times,
            plot_objs,
            where="post",
            linewidth=2,
            label=f"Seed {tr['seed']}"
        )
        ax.scatter(plot_times, plot_objs, s=25)
        plotted_any = True

ax.set_xlabel("Time (s)")
ax.set_ylabel("Objective Value")
ax.set_title("Phase 2: Objective Value Over Time from Warm Starts")
ax.grid(True, alpha=0.3)

if plotted_any:
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print("No heuristic solution updates were found in the logs.")
```

```python
# 1) keep prefer-resource constraints
cdf = df_constraints[df_constraints["constraint_type"] == "PreferResourcesConstraint"].copy()
cdf = cdf[cdf["required"].astype(str).str.upper() != "TRUE"]
cdf = cdf[~cdf["name"].astype(str).str.startswith("PreferResourceConstraint_Dept")]

cids = set(cdf["constraint_id"])
cid_to_weight = cdf.set_index("constraint_id")["weight"].astype(float).to_dict()

# 2) role per constraint (we only enforce Room-role constraints)
role_rows = df_constraint_params[
    (df_constraint_params["constraint_id"].isin(cids)) &
    (df_constraint_params["path"] == "Role")
][["constraint_id", "value"]]
cid_to_role = role_rows.set_index("constraint_id")["value"].to_dict()

# 3) preferred resources per constraint: direct resources
pref_resources = {cid: set() for cid in cids}

direct_res = df_constraint_params[
    (df_constraint_params["constraint_id"].isin(cids)) &
    (df_constraint_params["path"] == "Resources/Resource")
][["constraint_id", "value"]]
for r in direct_res.itertuples(index=False):
    pref_resources[r.constraint_id].add(r.value)


```

```python
# 4) preferred resources via resource groups
rg_rows = df_constraint_params[
    (df_constraint_params["constraint_id"].isin(cids)) &
    (df_constraint_params["path"] == "ResourceGroups/ResourceGroup")
][["constraint_id", "value"]]

# map roomgroup -> resources
rg_to_resources = (
    resource_groups.groupby("resource_group_id")["resource_id"].apply(set).to_dict()
)

for r in rg_rows.itertuples(index=False):
    pref_resources[r.constraint_id].update(rg_to_resources.get(r.value, set()))

# 5) applies-to events: direct + via event groups

# map constraint -> event applied to (one at a time)
applied_event = (
    df_constraint_applied[
        (df_constraint_applied["constraint_id"].isin(cids)) &
        (df_constraint_applied["path"] == "AppliesTo/Events/Event")
    ][["constraint_id", "reference"]]
    .groupby("constraint_id")["reference"].apply(set).to_dict()
)

# map eventgroup -> event 
applies_eg = df_constraint_applied[
    (df_constraint_applied["constraint_id"].isin(cids)) &
    (df_constraint_applied["path"] == "AppliesTo/EventGroups/EventGroup")
][["constraint_id", "reference"]]

eg_to_events = (
    df_event_groups_membership.groupby("event_group_id")["event_id"].apply(set).to_dict()
)

# map constraint -> event applied to (lists)
applied_events = {cid: set(applied_event.get(cid, set())) for cid in cids}
for r in applies_eg.itertuples(index=False):
    applied_events[r.constraint_id].update(eg_to_events.get(r.reference, set()))
```

```python
# 6) parent -> sub-events, sub-event -> duration
parent_to_sub = df_sub_events.groupby("parent_event_id")["sub_event_id"].apply(list).to_dict()
sub_duration = df_sub_events.set_index("sub_event_id")["sub_duration"].astype(int).to_dict()

all_rooms = set(df_rooms["resource_id"].tolist())

# 7) penalty expression
penalty = 0.0

for cid in cids: # for each constraint
    # only room-role prefer constraints in this function
    w = 1 # constraint weight
    pref = pref_resources.get(cid, set()) # constraint preferred resources
    nonpref = all_rooms - pref # constraint not preferred resources
    
    for pe in applied_events.get(cid, set()): # for each event in constraint
        for se in parent_to_sub.get(pe, []): # for each sub event in event
            d = int(sub_duration[se]) # duration of sub event
            # deviation contribution: duration if assigned to non-preferred room
            penalty += w * d * sum(m2.x[se, r] for r in nonpref)


m2.PreferRoomPenalty = pyo.Expression(expr=penalty)
```

```python
# 4) PreferResourcesConstraint_Dept
# Selected event prefers rooms from selected room-group(s)
# Soft, linear, weight = 1

# --- keep only dept prefer-resource constraints ---
cdf_dept = df_constraints[df_constraints["constraint_type"] == "PreferResourcesConstraint"].copy()
cdf_dept = cdf_dept[cdf_dept["required"].astype(str).str.upper() != "TRUE"]
cdf_dept = cdf_dept[cdf_dept["name"].astype(str).str.startswith("PreferResourceConstraint_Dept_")]

dept_cids = set(cdf_dept["constraint_id"])

# --- role map (optional safety) ---
role_rows = df_constraint_params[
    (df_constraint_params["constraint_id"].isin(dept_cids)) &
    (df_constraint_params["path"] == "Role")
][["constraint_id", "value"]]
cid_to_role = role_rows.set_index("constraint_id")["value"].to_dict()

# --- preferred resources per dept-constraint ---
pref_resources_dept = {cid: set() for cid in dept_cids}

# df constraint id and room group
rg_rows = df_constraint_params[
    (df_constraint_params["constraint_id"].isin(dept_cids)) &
    (df_constraint_params["path"] == "ResourceGroups/ResourceGroup")
][["constraint_id", "value"]]

# map resource group -> resource
rg_to_resources = (
    resource_groups.groupby("resource_group_id")["resource_id"].apply(set).to_dict()
)


for r in rg_rows.itertuples(index=False):
    pref_resources_dept[r.constraint_id].update(rg_to_resources.get(r.value, set()))

# --- applies-to events (direct + via event groups) ---

# map constraint -> event
applied_event = (
    df_constraint_applied[
        (df_constraint_applied["constraint_id"].isin(dept_cids)) &
        (df_constraint_applied["path"] == "AppliesTo/Events/Event")
    ][["constraint_id", "reference"]]
    .groupby("constraint_id")["reference"].apply(set).to_dict()
)

applies_eg = df_constraint_applied[
    (df_constraint_applied["constraint_id"].isin(dept_cids)) &
    (df_constraint_applied["path"] == "AppliesTo/EventGroups/EventGroup")
][["constraint_id", "reference"]]

# map event group -> events
eg_to_events = (
    df_event_groups_membership.groupby("event_group_id")["event_id"].apply(set).to_dict()
)

# map 
applied_events = {cid: set(applied_event.get(cid, set())) for cid in dept_cids}
for r in applies_eg.itertuples(index=False):
    applied_events[r.constraint_id].update(eg_to_events.get(r.reference, set()))

# --- parent -> sub-events / sub-event duration ---
parent_to_sub = df_sub_events.groupby("parent_event_id")["sub_event_id"].apply(list).to_dict()
sub_duration = df_sub_events.set_index("sub_event_id")["sub_duration"].astype(int).to_dict()

all_rooms = set(df_rooms["resource_id"].tolist())

# --- penalty expression ---
dept_penalty = 0.0

for cid in dept_cids: # for each constraint
    w = 1.0 # constraint weight
    pref = pref_resources_dept.get(cid, set()) # constraint preferred resources
    nonpref = all_rooms - pref # constraint not preferred rooms

    for pe in applied_events.get(cid, set()): # for event in constraint group
        for se in parent_to_sub.get(pe, []): # for sub event in event
            d = int(sub_duration[se]) # sub event duration
            dept_penalty += w * d * sum(m2.x[se, r] for r in nonpref)

m2.PreferDeptPenalty = pyo.Expression(expr=dept_penalty)
```

### Solve

```python
# ---- Objective ---- (with soft constraints)
penalty_total = (m2.PreferRoomPenalty + m2.PreferDeptPenalty)

m2.obj = pyo.Objective(expr=penalty_total, sense=pyo.minimize)
```
