### 2a: Model setup and helpers

```python
def as_list(x):
    if x is None:
        return []
    if isinstance(x, (list, tuple, set)):
        return list(x)
    return [x]
```

```python
### Sets

E = df_events["event_id"].tolist() # All events
T = df_times["time_id"].tolist() # All timeslots
R = df_rooms["resource_id"].tolist() # All rooms

all_teachers = df_teachers["resource_id"].tolist()
all_classes = df_classes["resource_id"].tolist()


# event -> duration/teachers/classes (lists)
event_to_duration = dict(zip(df_events["event_id"], df_events["duration"].astype(int))) # duration of each event
event_to_teachers = dict(zip(df_events["event_id"], df_events["teachers"])) # teachers for each event
event_to_classes  = dict(zip(df_events["event_id"], df_events["classes"])) # classes for each event


# event -> room group of each event
event_to_roomgroup = dict(zip(df_events["event_id"], df_events["role"])) 

# room -> room group of each room
room_groups = {"gr_Rooms_X", "gr_Rooms_Y", "gr_Rooms_Z"}
room_to_group = (
    df_rooms_long[df_rooms_long["resource_group"].isin(room_groups)]
    .set_index("resource_id")["resource_group"]
    .to_dict()
)

# room group -> rooms in that group
group_to_rooms = defaultdict(list)
for room, grp in room_to_group.items():   # room_to_group: room -> gr_Rooms_X/Y/Z
    group_to_rooms[grp].append(room)


T_index = {t:i for i,t in enumerate(T)}

# Unavailability rows for resources (teachers/classes/etc.)
df_unavail_times = df_constraint_params[
    (df_constraint_params["constraint_type"] == "AvoidUnavailableTimesConstraint") &
    (df_constraint_params["path"] == "Times/Time")
].copy()[["reference", "value"]]

# resource -> set(unavailable times)
resource_unavailability = (
    df_unavail_times
    .groupby("reference")["value"]
    .agg(lambda s: set(s))
    .to_dict()
)
```

```python
### Feasible Start times for events:

#T_index = {t:i for i,t in enumerate(T)}

time_to_day = dict(zip(df_times["time_id"], df_times["day_ref"]))

# Build feasible starts per event (no crossing day boundary) (implicitely covers constraints 5,9,10,11,12)
feasible_starts = {}
for e in E:
    d = int(event_to_duration[e]) # number of periods needed
    starts = []
    for ts in T: # for all possible start times
        i0 = T_index[ts]
        # must have enough periods remaining in the global list
        if i0 + d - 1 >= len(T):
            continue
        day0 = time_to_day[ts]
        block = T[i0:i0 + d]
        # enforce all times in the block are same day
        if all(time_to_day[t] == day0 for t in block):
            starts.append(ts)
    feasible_starts[e] = starts

# If event e starts at ts, then occ_times[(e, ts)] = [t1, t2, ...] is the list of all times occupied by that event
occ_times = {}
for e in E:
    d = int(event_to_duration[e])
    for ts in feasible_starts[e]:
        i0 = T_index[ts]
        occ_times[(e, ts)] = T[i0:i0 + d]


# Helper: all (e, ts) pairs that occupy a given time t
# ie for Fr_3, if event 1 is duration 1 and event 2 is duration 2, then (1, Fr_3) and (2, Fr_2) and (2, Fr_3) occupy it
occupies_at_time = {t: [] for t in T}
for e in E:
    for ts in feasible_starts[e]:
        for t in occ_times[(e, ts)]:
            occupies_at_time[t].append((e, ts))

```

```python

# Filter feasible starts against both teacher and class unavailability
feasible_starts_filtered = {}

for e in E:
    starts_ok = []
    teachers = set(as_list(event_to_teachers[e]))
    classes  = set(as_list(event_to_classes[e]))
    resources_for_event = teachers | classes

    for ts in feasible_starts[e]:
        occ = occ_times[(e, ts)]
        violates = False

        for res in resources_for_event:
            unavail = resource_unavailability.get(res, set())
            if any(t in unavail for t in occ):
                violates = True
                break

        if not violates:
            starts_ok.append(ts)

    feasible_starts_filtered[e] = starts_ok

# Replace
feasible_starts = feasible_starts_filtered
#feasible_starts


# rebuild occ_times
occ_times = {}
for e in E:
    d = int(event_to_duration[e])
    for ts in feasible_starts[e]:
        i0 = T_index[ts]
        occ_times[(e, ts)] = T[i0:i0 + d]

# rebuild occupies_at_time
occupies_at_time = {t: [] for t in T}
for e in E:
    for ts in feasible_starts[e]:
        for t in occ_times[(e, ts)]:
            occupies_at_time[t].append((e, ts))

feasible_starts
```

### 2b: Pyomo Model part 1: Sets and Decision Variables

Constraints:

Part 1: Event -> time:

1) AssignTimes: [gr_EventsGeneral] Every event must be assigned exactly one time
2) Do Not Split Events: [gr_EventsGeneral] Events must inhabit consecutive periods
3) PreferredTimesDurationTwo: [gr_EventsDuration2] Events with duration 2 must only be allocated to times in groups gr_TimesDurationTwo
4) EventSpreadingConstraint: [event groups] events from each goup have a maximum and minimum specified number of events in each time group [monday, tuesday etcc...] (here max 1 per day) (event group given in course ID) (soft: weight = 1, linear)
5) NoResourceClashes: No clashes for any of the resources
6) Unavailabilities_teacher: [selected teacher] selected Teacher cannot be allocated event at any of the given times
7) MinMaxADay_Classes: [Select Classes] Classes should attend between min and max periods every day (soft: weight = 1 quadratic)
8) MinMaxADay_Teachers: [Select Teachers] Teachers should attend between min and max periods every day (soft: weight = 1 quadratic)

Extra: Number of events at any time should be less than the number of total rooms 

Part 2: Event -> room:

1) Every event must be assigned exactly 1 room
2) At most one event can occupy a room at a time

### 2b.1: Model 1

```python
# ---- Model ----
m1 = pyo.ConcreteModel()

# ---- Sets ----
m1.E = pyo.Set(initialize=E) # Set of events
m1.T = pyo.Set(initialize=T, ordered=True) # Set of timeslots (ordered)
m1.R = pyo.Set(initialize=R) # Set of rooms

# standard set up
m1.X_index = pyo.Set(
    dimen=2,
    initialize=[(e, ts) for e in E for ts in feasible_starts[e]]
)



# ---- Decision Variables ----
# Start-time decision variables:
# x[e, ts, r] = 1 if event e starts at time ts in room r
# Only define variables for feasible starts (and all rooms) (THIS SPEEDS UP MODEL)
m1.x = pyo.Var(m1.X_index, domain=pyo.Binary)
```

```python

# 1) Each event scheduled exactly once (one start time)   
def one_start_rule(m, e):
    return sum(m.x[e, ts] for ts in feasible_starts[e]) == 1
m1.OneStart = pyo.Constraint(m1.E, rule=one_start_rule)


# 3) No teacher conflicts: a teacher can teach at most one event at a time
def teacher_conflict_rule(m, teacher, t):
    terms = [
        m.x[e, ts]
        for (e, ts) in occupies_at_time.get(t, [])
        if teacher in as_list(event_to_teachers[e])
    ]
    if not terms:
        return pyo.Constraint.Skip   # or pyo.Constraint.Feasible
    return sum(terms) <= 1

m1.TeacherConflict = pyo.Constraint(all_teachers, m1.T, rule=teacher_conflict_rule)


# 4) No class conflicts: a class can attend at most one event at a time
def class_conflict_rule(m, cls, t):
    terms = [
        m.x[e, ts]
        for (e, ts) in occupies_at_time.get(t, [])
        if cls in event_to_classes[e]
    ]
    if not terms:
        return pyo.Constraint.Skip
    return sum(terms) <= 1

m1.ClassConflict = pyo.Constraint(all_classes, m1.T, rule=class_conflict_rule)

# 5) Max number of simultaneous events cannot exceed number of rooms
m1.N_rooms = pyo.Param(initialize=len(R), within=pyo.NonNegativeIntegers)

def total_events_vs_rooms_rule(m, t):
    terms = [m.x[e, ts] for (e, ts) in occupies_at_time.get(t, [])]
    if not terms:
        return pyo.Constraint.Skip
    return sum(terms) <= m.N_rooms

m1.TotalEventsVsRooms = pyo.Constraint(m1.T, rule=total_events_vs_rooms_rule)
```

### Solve

```python
# ---- Objective ---- (hard constraints only)
m1.obj = pyo.Objective(expr=0, sense=pyo.minimize)
```

```python
seeds = [10, 20, 30, 40, 50]

solve_times = []
objectives = []
num_vars = []
num_cons = []
solve_status = []
start_x_list = []

import time as time_module

for seed in seeds:
    solver = pyo.SolverFactory("gurobi")
    solver.options["TimeLimit"] = 600
    solver.options["Seed"] = seed
    solver.options["Threads"] = 1

    # Manual timing
    start_time = time_module.time()
    res = solver.solve(m1, tee=False)
    elapsed_time = time_module.time() - start_time

    solve_times.append(elapsed_time)
    objectives.append(pyo.value(m1.obj))
    num_vars.append(m1.nvariables())
    num_cons.append(m1.nconstraints())
    solve_status.append(res.solver.termination_condition)

    # Store solution
    seed_solution = {}
    for idx in m1.X_index:
        val = pyo.value(m1.x[idx], exception=False)
        seed_solution[idx] = 0.0 if val is None else float(val)
    start_x_list.append(seed_solution)

# Keep compatibility with later cells that expect start_x
start_x = start_x_list[0]

# Calculate averages
avg_time = sum(solve_times) / len(solve_times)
avg_obj = sum(objectives) / len(objectives)
avg_vars = sum(num_vars) / len(num_vars)
avg_cons = sum(num_cons) / len(num_cons)

print("=" * 50)
print("SOLVER METRICS SUMMARY - PHASE 1")
print("=" * 50)
print(f"Average Solve Time (s): {avg_time:.4f}")
print(f"Average Objective: {avg_obj:.6f}")
print(f"Average Variables: {avg_vars:.0f}")
print(f"Average Constraints: {avg_cons:.0f}")
print(f"Termination Status: {set(solve_status)}")
print("=" * 50)

print(f"\nPhase 1 produced {len(start_x_list)} solutions (one per seed)")
for i, seed_sol in enumerate(start_x_list):
    scheduled = sum(1 for v in seed_sol.values() if v > 0)
    print(f"  Seed {seeds[i]}: {scheduled} events scheduled in {solve_times[i]:.2f}s, Obj={objectives[i]:.6f}")
```

### 2c: Model 2

```python
###  Events that are part of the same course must be on different days (soft) 


# Maps
# event -> course
event_to_course = dict(zip(df_events["event_id"], df_events["course_reference"]))

# spread constraint ids
spread_cons = df_constraints[
    df_constraints["constraint_type"] == "SpreadEventsConstraint"
]["constraint_id"].tolist()
if not spread_cons:
    raise ValueError("No SpreadEventsConstraint found in df_constraints.")

# Use AppliesTo from constraint_applies_to.csv
# (your notebook variable is df_constraint_applied)
spread_id = spread_cons[0]

spread_course_refs = set(
    df_constraint_applied[
        (df_constraint_applied["constraint_id"] == spread_id) &
        (df_constraint_applied["path"] == "AppliesTo/EventGroups/EventGroup")
    ]["reference"].dropna().tolist()
)

# Restrict courses to only those in AppliesTo and present in events
all_event_courses = set(event_to_course[e] for e in E)
courses = sorted(all_event_courses.intersection(spread_course_refs))
days = sorted(df_times["day_ref"].dropna().unique().tolist())

# If model uses m instead of m1, rename below accordingly
m1.COURSES = pyo.Set(initialize=courses)
if not hasattr(m1, "DAYS"):
    m1.DAYS = pyo.Set(initialize=days)

# ------------------------------------------------------------
# 2) Read SpreadEvents min/max by day from constraint data
#    Expected rows in df_constraint_params:
# path
spread_cons = df_constraints[df_constraints["constraint_type"] == "SpreadEventsConstraint"]["constraint_id"].tolist()
if len(spread_cons) == 0:
    raise ValueError("No SpreadEventsConstraint found in df_constraints.")
spread_id = spread_cons[0]  # if multiple, choose/merge as needed

cp = df_constraint_params[df_constraint_params["constraint_id"] == spread_id].copy()

# build day->(min,max)
day_bounds = {}
refs = cp[cp["path"] == "TimeGroups/TimeGroup"][["reference"]].dropna().drop_duplicates()

for _, row in refs.iterrows():
    d = row["reference"]
    min_row = cp[(cp["path"] == "TimeGroups/TimeGroup/Minimum") & (cp["reference"] == d)]
    max_row = cp[(cp["path"] == "TimeGroups/TimeGroup/Maximum") & (cp["reference"] == d)]

    dmin = int(min_row["value"].iloc[0]) if len(min_row) else 0
    dmax = int(max_row["value"].iloc[0]) if len(max_row) else 1
    day_bounds[d] = (dmin, dmax)

# default for any day missing in XML params
for d in days:
    day_bounds.setdefault(d, (0, 1))

spread_min = {d: day_bounds[d][0] for d in days}
spread_max = {d: day_bounds[d][1] for d in days}

m1.spread_min = pyo.Param(m1.DAYS, initialize=spread_min, within=pyo.NonNegativeIntegers)
m1.spread_max = pyo.Param(m1.DAYS, initialize=spread_max, within=pyo.NonNegativeIntegers)

# Build
course_day_terms = defaultdict(list)
for (e, ts) in m1.X_index:
    c = event_to_course[e]
    d = time_to_day[ts]
    course_day_terms[(c, d)].append((e, ts))

# deviations
m1.spread_under = pyo.Var(m1.COURSES, m1.DAYS, domain=pyo.NonNegativeReals)
m1.spread_over  = pyo.Var(m1.COURSES, m1.DAYS, domain=pyo.NonNegativeReals)

def spread_min_soft_rule(m, c, d):
    load = sum(m.x[e, ts] for (e, ts) in course_day_terms.get((c, d), []))
    return load + m.spread_under[c, d] >= m.spread_min[d]
m1.SpreadMinSoft = pyo.Constraint(m1.COURSES, m1.DAYS, rule=spread_min_soft_rule)

def spread_max_soft_rule(m, c, d):
    load = sum(m.x[e, ts] for (e, ts) in course_day_terms.get((c, d), []))
    return load - m.spread_over[c, d] <= m.spread_max[d]
m1.SpreadMaxSoft = pyo.Constraint(m1.COURSES, m1.DAYS, rule=spread_max_soft_rule)

# Linear penalty, weight = 1
m1.PenaltySpreadEvents = pyo.Expression(
    expr=sum(
        m1.spread_under[c, d] + m1.spread_over[c, d]
        for c in m1.COURSES for d in m1.DAYS
    )
)
```

### MinMaxADay Classes

```python

def refs_for(constraint_id):
    return sorted(
        df_constraint_applied.loc[
            df_constraint_applied["constraint_id"].eq(constraint_id), "reference"
        ].dropna().unique().tolist()
    )


classes_1 = refs_for("MinMaxADay_Classes_1_mo")   # min=3, max=6
classes_2 = refs_for("MinMaxADay_Classes_2_mo")   # min=1, max=2
classes_3 = refs_for("MinMaxADay_Classes_3_mo")   # min=1, max=6
classes_4 = refs_for("MinMaxADay_Classes_4_mo")   # min=1, max=7
classes_5 = refs_for("MinMaxADay_Classes_5_mo")   # min=3, max=7

teachers_1 = all_teachers # min=2, max=7


print("classes_1:", classes_1)
print("classes_2:", classes_2)
print("classes_3:", classes_3)
print("classes_4:", classes_4)
print("classes_5:", classes_5)

print("teachers_1:", teachers_1)


def as_list(x):
    if x is None:
        return []
    if isinstance(x, (list, tuple, set)):
        return list(x)
    return [x]



# Requires: m.X_index, event_to_classes, event_to_teachers, time_to_day

class_day_terms = defaultdict(list)    # key: (class, day)   -> list[(e,ts)] (feasible combinations)
teacher_day_terms = defaultdict(list)  # key: (teacher, day) -> list[(e,ts)]  (feasible combinations)

for (e, ts) in m1.X_index:
    d = time_to_day[ts]

    for cls in as_list(event_to_classes[e]):
        class_day_terms[(cls, d)].append((e, ts))

    for tch in as_list(event_to_teachers[e]):
        teacher_day_terms[(tch, d)].append((e, ts))


# Build day -> timeslots once
day_to_times = defaultdict(list)
for t in m1.T:
    day_to_times[time_to_day[t]].append(t)
```

```python
m1.DAYS = pyo.Set(initialize=days)
```

```python
# 1) classes_1 (min=3, max=6)

C1_MIN, C1_MAX = 3, 6
m1.C1 = pyo.Set(initialize=list(classes_1))
#m1.c1_days = pyo.Set(initialize=list(m1.DAYS))

# terms[(cls,t)] = feasible x[e,ts] making class busy at t
c1_terms = defaultdict(list) # for each [(cls,t)], all the feasible (event,timestart) combos that would make class busy at that time
c1_set = set(classes_1) # set of classes in classes_1
for t in m1.T:
    for (e, ts) in occupies_at_time.get(t, []):
        for cls in as_list(event_to_classes[e]):
            if cls in c1_set:
                c1_terms[(cls, t)].append((e, ts))

m1.c1_present = pyo.Var(m1.C1, m1.DAYS, domain=pyo.Binary) # =1 if class is busy at least once on day d
m1.c1_under   = pyo.Var(m1.C1, m1.DAYS, domain=pyo.NonNegativeReals) # classes_1 shortage of events below minimum (1)
m1.c1_over    = pyo.Var(m1.C1, m1.DAYS, domain=pyo.NonNegativeReals) # classes_1 excess of events above maximum (1)

# Iterates over each timeslot in a day and for each t, it adds events that occupy that t (from c1_terms[(cls, t)] (ie. calculates number of busy times in the day)
def c1_busy_rule(m, cls, d):
    return sum(m.x[e, ts] for t in day_to_times[d] for (e, ts) in c1_terms.get((cls, t), []))
m1.c1_busy = pyo.Expression(m1.C1, m1.DAYS, rule=c1_busy_rule)


# Together these force c1_present = 0 if c1_busy = 0 and c1_present = 1 if c1_busy >= 1
def c1_present_upper(m, cls, d):
    return m.c1_busy[cls, d] <= len(day_to_times[d]) * m.c1_present[cls, d]
m1.C1PresentUpper = pyo.Constraint(m1.C1, m1.DAYS, rule=c1_present_upper)

def c1_present_lower(m, cls, d):
    return m.c1_busy[cls, d] >= m.c1_present[cls, d]
m1.C1PresentLower = pyo.Constraint(m1.C1, m1.DAYS, rule=c1_present_lower)


# Measures shortage below the minimum, class not present on a day -> RHS is -c1_busy -> busy forced to 0. Presnt = 1 and busy < C1-min -> RHS is +ve and under must cover gap
def c1_under_def(m, cls, d):
    return m.c1_under[cls, d] >= C1_MIN * m.c1_present[cls, d] - m.c1_busy[cls, d]
m1.C1UnderDef = pyo.Constraint(m1.C1, m1.DAYS, rule=c1_under_def)

# Measures exess above maximum. busy > C1_Max -> RHS +ve so over must cover exess etc...
def c1_over_def(m, cls, d):
    return m.c1_over[cls, d] >= m.c1_busy[cls, d] - C1_MAX
m1.C1OverDef = pyo.Constraint(m1.C1, m1.DAYS, rule=c1_over_def)

m1.PenaltyClass1Busy = pyo.Expression(
    expr=sum((m1.c1_under[c,d] + m1.c1_over[c,d])**2 for c in m1.C1 for d in m1.DAYS) # squares total deviation for each class-day
)
```

```python
import os
import re
import time as time_module
import numpy as np
import matplotlib.pyplot as plt
import pyomo.environ as pyo

TIME_LIMIT = 600
LOG_DIR = "gurobi_phase2_logs"
SENSE = "min"
os.makedirs(LOG_DIR, exist_ok=True)


def clear_all_var_values(model):
    for v in model.component_data_objects(pyo.Var, active=True):
        v.set_value(None, skip_validation=True)


def load_x_warm_start(model, seed_solution, tol=0.5):
    # Start from a clean model state so no stale values are exported
    clear_all_var_values(model)

    assigned = 0
    for idx in model.X_index:
        val = seed_solution.get(idx, 0.0)
        start_val = 1 if val > tol else 0
        model.x[idx].set_value(start_val, skip_validation=True)
        assigned += 1

    return assigned


def parse_model_stats(logfile):
    rows = cols = nnz = None
    pat = re.compile(
        r"Optimize a model with\s+(\d+)\s+rows,\s+(\d+)\s+columns and\s+(\d+)\s+nonzeros"
    )
    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            m = pat.search(line)
            if m:
                rows = int(m.group(1))
                cols = int(m.group(2))
                nnz = int(m.group(3))
                break
    return rows, cols, nnz


def parse_heuristic_trajectory(logfile):
    times = []
    objs = []

    elapsed_pat = re.compile(
        r"Elapsed time for NoRel heuristic:\s*([0-9]+(?:\.[0-9]+)?)s"
    )
    found_pat = re.compile(
        r"Found heuristic solution:\s*objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )
    mip_start_pat = re.compile(
        r"Loaded user MIP start with objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )

    current_time = None

    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()

            m = elapsed_pat.search(s)
            if m:
                current_time = float(m.group(1))
                continue

            m = mip_start_pat.search(s)
            if m:
                times.append(0.0)
                objs.append(float(m.group(1)))
                continue

            m = found_pat.search(s)
            if m:
                times.append(current_time)
                objs.append(float(m.group(1)))
                continue

            if s.startswith("H") or s.startswith("*"):
                toks = s.split()
                if toks[0] == "H" and len(toks) >= 8:
                    objs.append(float(toks[3]))
                    times.append(float(toks[-1].rstrip("s")))
                elif toks[0] == "*" and len(toks) >= 9:
                    objs.append(float(toks[4]))
                    times.append(float(toks[-1].rstrip("s")))

    return times, objs



def parse_first_feasible_time(logfile):
    elapsed_pat = re.compile(
        r"Elapsed time for NoRel heuristic:\s*([0-9]+(?:\.[0-9]+)?)s"
    )
    found_pat = re.compile(
        r"Found heuristic solution:\s*objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )
    mip_start_pat = re.compile(
        r"Loaded user MIP start with objective\s*([+-]?\d+(?:\.\d+)?(?:[eE][+-]?\d+)?)"
    )

    current_time = None

    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()

            m = elapsed_pat.search(s)
            if m:
                current_time = float(m.group(1))
                continue

            m = mip_start_pat.search(s)
            if m:
                return 0.0, float(m.group(1)), "warm_start"

            m = found_pat.search(s)
            if m:
                return current_time, float(m.group(1)), "heuristic_log"

    return None, None, None


def parse_mip_start_status(logfile):
    accepted = False
    messages = []

    with open(logfile, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()
            if "User MIP start" in s or "Loaded user MIP start" in s:
                messages.append(s)
            if "Loaded user MIP start" in s or "User MIP start produced solution" in s:
                accepted = True

    return accepted, messages


phase2_results = []
trajectories = []

for i, seed_solution in enumerate(start_x_list):
    seed = seeds[i]

    print(f"\n{'='*60}")
    print(f"PHASE 2 - Warm Start from Seed {seed}")
    print(f"{'='*60}")

    assigned = load_x_warm_start(m1, seed_solution)

    logfile = os.path.join(LOG_DIR, f"phase2_seed_{seed}.log")
    if os.path.exists(logfile):
        os.remove(logfile)

    solver2 = pyo.SolverFactory("gurobi")
    solver2.options["TimeLimit"] = TIME_LIMIT
    solver2.options["Seed"] = seed
    solver2.options["MIPFocus"] = 1
    solver2.options["Heuristics"] = 0.5
    solver2.options["NoRelHeurTime"] = TIME_LIMIT
    solver2.options["DisplayInterval"] = 1
    solver2.options["LogFile"] = logfile
    solver2.options["Threads"] = 1

    start_time = time_module.time()
    res2 = solver2.solve(m1, tee=False, warmstart=True, load_solutions=False)
    elapsed_time = time_module.time() - start_time

    solver_status = str(res2.solver.status)
    termination = str(res2.solver.termination_condition)

    has_solution = len(res2.solution) > 0
    if has_solution:
        m1.solutions.load_from(res2)

    try:
        obj2 = pyo.value(m1.obj) if has_solution else float("nan")
    except:
        obj2 = float("nan")
        has_solution = False

    chosen_start = {}
    if has_solution:
        for e in m1.E:
            picked = [
                ts for ts in feasible_starts[e]
                if pyo.value(m1.x[e, ts], exception=False) is not None
                and pyo.value(m1.x[e, ts]) > 0.5
            ]
            if len(picked) != 1:
                raise ValueError(f"Event {e}: expected exactly 1 chosen start, got {picked}")
            chosen_start[e] = picked[0]


    num_vars = m1.nvariables()
    num_cons = m1.nconstraints()
    rows_log, cols_log, nnz_log = parse_model_stats(logfile)
    tvals, ovals = parse_heuristic_trajectory(logfile)
    first_feas_time, first_feas_obj, first_feas_source = parse_first_feasible_time(logfile)
    mip_start_accepted, mip_start_messages = parse_mip_start_status(logfile)

    if has_solution:
        if not tvals:
            tvals.append(elapsed_time)
            ovals.append(obj2)
        elif abs(ovals[-1] - obj2) > 1e-9 and elapsed_time > (tvals[-1] if tvals[-1] is not None else -1):
            tvals.append(elapsed_time)
            ovals.append(obj2)

    phase2_results.append({
        "seed": seed,
        "solver_status": solver_status,
        "termination": termination,
        "objective": obj2,
        "solve_time": elapsed_time,
        "num_vars": num_vars if num_vars is not None else cols_log,
        "num_cons": num_cons if num_cons is not None else rows_log,
        "num_nonzeros": nnz_log,
        "has_solution": has_solution,
        "assigned_start_values": assigned,
        "chosen_start": chosen_start,
        "mip_start_accepted": mip_start_accepted,
        "mip_start_messages": mip_start_messages,
        "n_heuristic_updates": len(tvals),
        "first_feasible_time": first_feas_time,
        "first_feasible_objective": first_feas_obj,
        "first_feasible_source": first_feas_source,
        "logfile": logfile,
    })


    trajectories.append({
        "seed": seed,
        "times": tvals,
        "objectives": ovals,
    })

    print(f"Assigned x warm-start values: {assigned}")
    print(f"Solver status: {solver_status}")
    print(f"Termination: {termination}")
    print(f"Warm start accepted: {mip_start_accepted}")
    if mip_start_messages:
        print("MIP start log:")
        for msg in mip_start_messages:
            print("  ", msg)
    print(f"Objective: {obj2:.6f}" if has_solution else "Objective: No solution loaded")
    print(f"Solve time: {elapsed_time:.2f}s")
    print(f"Heuristic updates found: {len(tvals)}")
    print(f"Variables: {num_vars}")
    print(f"Constraints: {num_cons}")
    print(f"Nonzeros: {nnz_log}")


# SUMMARY
valid_objs = [r["objective"] for r in phase2_results if not np.isnan(r["objective"])]
avg_time = np.mean([r["solve_time"] for r in phase2_results])

print(f"\n{'='*60}")
print("PHASE 2 SUMMARY")
print(f"{'='*60}")
for r in phase2_results:
    obj_str = f"{r['objective']:.6f}" if not np.isnan(r["objective"]) else "No solution"
    first_time_str = (
        f"{r['first_feasible_time']:.2f}s"
        if r["first_feasible_time"] is not None
        else "Not found"
    )
    first_obj_str = (
        f"{r['first_feasible_objective']:.6f}"
        if r["first_feasible_objective"] is not None
        else "Not found"
    )
    print(
        f"Seed {r['seed']}: "
        f"Status={r['solver_status']}, "
        f"Termination={r['termination']}, "
        f"Obj={obj_str}, "
        f"Time={r['solve_time']:.2f}s, "
        f"FirstFeasibleTime={first_time_str}, "
        f"FirstFeasibleObj={first_obj_str}, "
        f"FirstFeasibleSource={r['first_feasible_source']}, "
        f"HeuristicUpdates={r['n_heuristic_updates']}"
    )

print(f"\nAverage Solve Time (s): {avg_time:.4f}")
print(f"Average Objective: {np.mean(valid_objs):.6f}" if valid_objs else "Average Objective: Not available")

feasible_results = [r for r in phase2_results if not np.isnan(r["objective"])]
if feasible_results:
    best_result = min(feasible_results, key=lambda x: x["objective"]) if SENSE == "min" else max(feasible_results, key=lambda x: x["objective"])
    print(
        f"Best Phase 2 Solution: Seed {best_result['seed']} "
        f"with objective {best_result['objective']:.6f}"
    )
else:
    print("Best Phase 2 Solution: No feasible solution was loaded")


# PLOT
fig, ax = plt.subplots(figsize=(10, 6))

plotted_any = False
for tr in trajectories:
    plot_times = [0.0 if t is None else t for t in tr["times"]]
    plot_pairs = sorted(zip(plot_times, tr["objectives"]), key=lambda x: x[0])

    if plot_pairs:
        plot_times, plot_objs = zip(*plot_pairs)

        ax.step(
            plot_times,
            plot_objs,
            where="post",
            linewidth=2,
            label=f"Seed {tr['seed']}"
        )
        ax.scatter(plot_times, plot_objs, s=25)
        plotted_any = True

ax.set_xlabel("Time (s)")
ax.set_ylabel("Objective Value")
ax.set_title("Phase 2: Objective Value Over Time from Warm Starts")
ax.grid(True, alpha=0.3)

if plotted_any:
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print("No heuristic solution updates were found in the logs.")
```

```python
def _as_list(x):
    if x is None:
        return []
    if isinstance(x, (list, tuple, set)):
        return list(x)
    return [x]


# Step

def extract_selected_m1(m1, tol=0.5):
    sel = []
    for (e, ts) in m1.X_index:
        v = pyo.value(m1.x[e, ts], exception=False)
        if v is not None and v > tol:
            sel.append((e, ts))
    return sel

def chosen_start_from_m1(m1, feasible_starts, tol=0.5):
    chosen = {}
    for e in m1.E:
        picks = [ts for ts in feasible_starts[e] if pyo.value(m1.x[e, ts], exception=False) is not None and pyo.value(m1.x[e, ts]) > tol]
        if len(picks) != 1:
            raise ValueError(f"Event {e}: expected exactly one chosen start, got {picks}")
        chosen[e] = picks[0]
    return chosen

def check_m1_one_start(m1, E):
    sel = extract_selected_m1(m1)
    by_e = defaultdict(int)
    for e, ts in sel:
        by_e[e] += 1
    missing = [e for e in E if by_e[e] == 0]
    multi = [e for e, c in by_e.items() if c > 1]
    return {"ok": len(missing) == 0 and len(multi) == 0, "missing": missing, "multiple": multi}

def check_m1_teacher_conflicts(m1, T, occupies_at_time, event_to_teachers):
    sel = set(extract_selected_m1(m1))
    bad = []
    for t in T:
        load = defaultdict(list)  # teacher -> [(e,ts)]
        for (e, ts) in occupies_at_time.get(t, []):
            if (e, ts) in sel:
                for teacher in _as_list(event_to_teachers[e]):
                    load[teacher].append((e, ts))
        for teacher, assigns in load.items():
            if len(assigns) > 1:
                bad.append((teacher, t, assigns))
    return {"ok": len(bad) == 0, "violations": bad}

def check_m1_class_conflicts(m1, T, occupies_at_time, event_to_classes):
    sel = set(extract_selected_m1(m1))
    bad = []
    for t in T:
        load = defaultdict(list)  # class -> [(e,ts)]
        for (e, ts) in occupies_at_time.get(t, []):
            if (e, ts) in sel:
                for cls in _as_list(event_to_classes[e]):
                    load[cls].append((e, ts))
        for cls, assigns in load.items():
            if len(assigns) > 1:
                bad.append((cls, t, assigns))
    return {"ok": len(bad) == 0, "violations": bad}

def check_m1_roomgroup_capacity(m1, T, occupies_at_time, event_to_roomgroup, roomgroup_cap):
    sel = set(extract_selected_m1(m1))
    bad = []
    for t in T:
        grp_load = defaultdict(list)  # group -> [(e,ts)]
        for (e, ts) in occupies_at_time.get(t, []):
            if (e, ts) in sel:
                g = event_to_roomgroup[e]
                grp_load[g].append((e, ts))
        for g, assigns in grp_load.items():
            cap = roomgroup_cap[g]
            if len(assigns) > cap:
                bad.append((g, t, len(assigns), cap, assigns))
    return {"ok": len(bad) == 0, "violations": bad}

def check_m1_event_spread_by_day(m1, event_to_course, time_to_day, day_max=1):
    sel = extract_selected_m1(m1)
    load = defaultdict(list)  # (course, day) -> [(e,ts)]
    for e, ts in sel:
        c = event_to_course[e]
        d = time_to_day[ts]
        load[(c, d)].append((e, ts))
    bad = [(c, d, assigns) for (c, d), assigns in load.items() if len(assigns) > day_max]
    return {"ok": len(bad) == 0, "violations": bad}

def run_m1_sanity_checks(
    m1, E, T, occupies_at_time, event_to_teachers, event_to_classes,
    event_to_course, time_to_day, event_to_roomgroup, roomgroup_cap
):
    out = {
        "OneStart": check_m1_one_start(m1, E),
        "TeacherConflict": check_m1_teacher_conflicts(m1, T, occupies_at_time, event_to_teachers),
        "ClassConflict": check_m1_class_conflicts(m1, T, occupies_at_time, event_to_classes),
        "RoomGroupCapacity": check_m1_roomgroup_capacity(m1, T, occupies_at_time, event_to_roomgroup, roomgroup_cap),
        "EventSpreadByDay": check_m1_event_spread_by_day(m1, event_to_course, time_to_day, day_max=1),
    }
    summary = pd.DataFrame([
        {
            "check": k,
            "ok": v["ok"],
            "n_violations": len(v.get("violations", [])) + len(v.get("missing", [])) + len(v.get("multiple", []))
        }
        for k, v in out.items()
    ])
    return out, summary


# Step

def extract_selected_m2(m2, tol=0.5):
    sel = []
    for (e, r) in m2.X_index:
        v = pyo.value(m2.x[e, r], exception=False)
        if v is not None and v > tol:
            sel.append((e, r))
    return sel

def check_m2_one_room(m2, E):
    sel = extract_selected_m2(m2)
    by_e = defaultdict(int)
    for e, r in sel:
        by_e[e] += 1
    missing = [e for e in E if by_e[e] == 0]
    multi = [e for e, c in by_e.items() if c > 1]
    return {"ok": len(missing) == 0 and len(multi) == 0, "missing": missing, "multiple": multi}

def check_m2_room_conflicts(m2, T, chosen_start, occ_times):
    sel = set(extract_selected_m2(m2))
    bad = []
    for t in T:
        room_load = defaultdict(list)  # room -> [event]
        for (e, r) in sel:
            ts = chosen_start[e]
            if t in occ_times[(e, ts)]:
                room_load[r].append(e)
        for r, evs in room_load.items():
            if len(evs) > 1:
                bad.append((r, t, evs))
    return {"ok": len(bad) == 0, "violations": bad}

def run_m2_sanity_checks(m2, E, T, chosen_start, occ_times):
    out = {
        "OneRoom": check_m2_one_room(m2, E),
        "RoomConflict": check_m2_room_conflicts(m2, T, chosen_start, occ_times),
    }
    summary = pd.DataFrame([
        {
            "check": k,
            "ok": v["ok"],
            "n_violations": len(v.get("violations", [])) + len(v.get("missing", [])) + len(v.get("multiple", []))
        }
        for k, v in out.items()
    ])
    return out, summary
```

```python
# SANITY CHECKER

# ---------- helpers ----------
def as_list(x):
    if x is None:
        return []
    if isinstance(x, (list, tuple, set)):
        return list(x)
    return [x]

def val01(v, tol=0.5):
    vv = pyo.value(v, exception=False)
    if vv is None:
        return 0
    return 1 if vv > tol else 0

def extract_selected_starts(m1, tol=0.5):
    sel = set()
    for (e, ts) in m1.X_index:
        if val01(m1.x[e, ts], tol=tol) == 1:
            sel.add((e, ts))
    return sel

def _safe_value(expr, default=0.0):
    v = pyo.value(expr, exception=False)
    return default if v is None else float(v)

# ---------- explicit hard checks ----------
def check_one_start(m1, E, feasible_starts, tol=0.5):
    selected = extract_selected_starts(m1, tol=tol)
    by_e = defaultdict(int)
    for e, ts in selected:
        by_e[e] += 1

    missing = [e for e in E if by_e[e] == 0]
    multiple = [e for e, c in by_e.items() if c > 1]
    violations = [{"event": e, "type": "missing"} for e in missing] + \
                 [{"event": e, "type": "multiple", "count": by_e[e]} for e in multiple]

    ok = (len(violations) == 0)
    detail = pd.DataFrame(violations) if violations else pd.DataFrame(columns=["event", "type", "count"])
    return ok, detail, {"n_violations": len(violations)}

def check_teacher_conflicts(m1, T, occupies_at_time, event_to_teachers, tol=0.5):
    selected = extract_selected_starts(m1, tol=tol)
    bad = []

    for t in T:
        load = defaultdict(list)
        for (e, ts) in occupies_at_time.get(t, []):
            if (e, ts) in selected:
                for teacher in as_list(event_to_teachers[e]):
                    load[teacher].append((e, ts))
        for teacher, assigns in load.items():
            if len(assigns) > 1:
                bad.append({
                    "teacher": teacher,
                    "time": t,
                    "count": len(assigns),
                    "assignments": str(assigns),
                })

    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(columns=["teacher", "time", "count", "assignments"])
    return ok, detail, {"n_violations": len(bad)}

def check_class_conflicts(m1, T, occupies_at_time, event_to_classes, tol=0.5):
    selected = extract_selected_starts(m1, tol=tol)
    bad = []

    for t in T:
        load = defaultdict(list)
        for (e, ts) in occupies_at_time.get(t, []):
            if (e, ts) in selected:
                for cls in as_list(event_to_classes[e]):
                    load[cls].append((e, ts))
        for cls, assigns in load.items():
            if len(assigns) > 1:
                bad.append({
                    "class": cls,
                    "time": t,
                    "count": len(assigns),
                    "assignments": str(assigns),
                })

    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(columns=["class", "time", "count", "assignments"])
    return ok, detail, {"n_violations": len(bad)}

def check_total_events_vs_rooms(m1, T, R, occupies_at_time, tol=0.5):
    selected = extract_selected_starts(m1, tol=tol)
    n_rooms = len(R)
    bad = []

    for t in T:
        load = sum(1 for (e, ts) in occupies_at_time.get(t, []) if (e, ts) in selected)
        if load > n_rooms:
            bad.append({"time": t, "active_events": load, "n_rooms": n_rooms})

    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(columns=["time", "active_events", "n_rooms"])
    return ok, detail, {"n_violations": len(bad)}

# ---------- explicit soft checks ----------
def check_spread_soft(m1, event_to_course, time_to_day, tol=1e-6):
    selected = extract_selected_starts(m1, tol=0.5)

    load = defaultdict(int)  # (course, day) -> starts
    for (e, ts) in selected:
        c = event_to_course[e]
        d = time_to_day[ts]
        load[(c, d)] += 1

    rows = []
    for c in m1.COURSES:
        for d in m1.DAYS:
            l = load.get((c, d), 0)
            u = _safe_value(m1.spread_under[c, d])
            o = _safe_value(m1.spread_over[c, d])
            mn = int(_safe_value(m1.spread_min[d]))
            mx = int(_safe_value(m1.spread_max[d]))

            min_ok = (l + u >= mn - tol)
            max_ok = (l - o <= mx + tol)
            rows.append({
                "course": c, "day": d, "load": l,
                "under": u, "over": o, "min": mn, "max": mx,
                "min_ok": min_ok, "max_ok": max_ok
            })

    df = pd.DataFrame(rows)
    viol = df[(~df["min_ok"]) | (~df["max_ok"])].copy()
    model_pen = _safe_value(m1.PenaltySpreadEvents)
    recomputed_pen = float((df["under"] + df["over"]).sum())
    pen_diff = abs(model_pen - recomputed_pen)

    ok = (len(viol) == 0) and (pen_diff <= 1e-5)
    meta = {
        "n_violations": int(len(viol)),
        "model_penalty": model_pen,
        "recomputed_penalty": recomputed_pen,
        "penalty_abs_diff": pen_diff,
    }
    return ok, viol if len(viol) else df.head(0), meta

def check_minmax_group_soft(m1, set_name, busy_name, present_name, under_name, over_name,
                            min_val, max_val, penalty_name, tol=1e-6):
    RES = getattr(m1, set_name)
    busy = getattr(m1, busy_name)
    present = getattr(m1, present_name)
    under = getattr(m1, under_name)
    over = getattr(m1, over_name)

    rows = []
    for r in RES:
        for d in m1.DAYS:
            b = _safe_value(busy[r, d])
            y = _safe_value(present[r, d])
            u = _safe_value(under[r, d])
            o = _safe_value(over[r, d])

            present_up_ok = b <= len([t for t in m1.T if True]) * y + tol  # loose global bound check
            present_lo_ok = b + tol >= y
            under_ok = u + tol >= (min_val * y - b)
            over_ok = o + tol >= (b - max_val)

            rows.append({
                "resource": r, "day": d, "busy": b, "present": y,
                "under": u, "over": o,
                "present_up_ok": present_up_ok,
                "present_lo_ok": present_lo_ok,
                "under_ok": under_ok,
                "over_ok": over_ok,
            })

    df = pd.DataFrame(rows)
    viol = df[(~df["present_up_ok"]) | (~df["present_lo_ok"]) | (~df["under_ok"]) | (~df["over_ok"])].copy()

    model_pen = _safe_value(getattr(m1, penalty_name))
    recomputed_pen = float(((df["under"] + df["over"]) ** 2).sum())
    pen_diff = abs(model_pen - recomputed_pen)

    ok = (len(viol) == 0) and (pen_diff <= 1e-5)
    meta = {
        "n_violations": int(len(viol)),
        "model_penalty": model_pen,
        "recomputed_penalty": recomputed_pen,
        "penalty_abs_diff": pen_diff,
    }
    return ok, viol if len(viol) else df.head(0), meta

# ---------- implicit checks ----------
def check_implicit_feasible_starts(E, feasible_starts):
    bad = [{"event": e, "n_feasible_starts": len(feasible_starts.get(e, []))}
           for e in E if len(feasible_starts.get(e, [])) == 0]
    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(columns=["event", "n_feasible_starts"])
    return ok, detail, {"n_violations": len(bad)}

def check_implicit_unavailability(feasible_starts, occ_times, event_to_teachers, event_to_classes, resource_unavailability):
    bad = []
    for e, starts in feasible_starts.items():
        resources = set(as_list(event_to_teachers[e])) | set(as_list(event_to_classes[e]))
        for ts in starts:
            occ = occ_times[(e, ts)]
            for r in resources:
                unavail = resource_unavailability.get(r, set())
                hit = [t for t in occ if t in unavail]
                if hit:
                    bad.append({"event": e, "start": ts, "resource": r, "violating_times": str(hit)})
    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(columns=["event", "start", "resource", "violating_times"])
    return ok, detail, {"n_violations": len(bad)}

def check_implicit_duration_day(feasible_starts, occ_times, event_to_duration, time_to_day):
    bad = []
    for e, starts in feasible_starts.items():
        dur = int(event_to_duration[e])
        for ts in starts:
            occ = occ_times[(e, ts)]
            same_day = len({time_to_day[t] for t in occ}) == 1
            if len(occ) != dur or not same_day:
                bad.append({
                    "event": e, "start": ts, "duration_expected": dur,
                    "duration_actual": len(occ), "same_day": same_day, "occ": str(occ)
                })
    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(
        columns=["event", "start", "duration_expected", "duration_actual", "same_day", "occ"]
    )
    return ok, detail, {"n_violations": len(bad)}

def check_implicit_feasible_rooms(E, feasible_rooms):
    if feasible_rooms is None:
        return True, pd.DataFrame(columns=["event", "n_feasible_rooms"]), {"n_violations": 0}
    bad = [{"event": e, "n_feasible_rooms": len(feasible_rooms.get(e, []))}
           for e in E if len(feasible_rooms.get(e, [])) == 0]
    ok = (len(bad) == 0)
    detail = pd.DataFrame(bad) if bad else pd.DataFrame(columns=["event", "n_feasible_rooms"])
    return ok, detail, {"n_violations": len(bad)}

# ---------- main runner ----------
def run_all_sanity_checks_elementary(
    m1,
    E, T, R,
    feasible_starts, occ_times, occupies_at_time, time_to_day,
    event_to_duration, event_to_teachers, event_to_classes, event_to_course,
    resource_unavailability,
    feasible_rooms=None,
    tol=1e-6
):
    summary = []
    explicit_details = {}
    implicit_details = {}

    # Hard explicit
    ok, det, meta = check_one_start(m1, E, feasible_starts)
    summary.append({"check": "OneStart", "category": "hard_explicit", "ok": ok, **meta})
    explicit_details["OneStart"] = det

    ok, det, meta = check_teacher_conflicts(m1, T, occupies_at_time, event_to_teachers)
    summary.append({"check": "TeacherConflict", "category": "hard_explicit", "ok": ok, **meta})
    explicit_details["TeacherConflict"] = det

    ok, det, meta = check_class_conflicts(m1, T, occupies_at_time, event_to_classes)
    summary.append({"check": "ClassConflict", "category": "hard_explicit", "ok": ok, **meta})
    explicit_details["ClassConflict"] = det

    ok, det, meta = check_total_events_vs_rooms(m1, T, R, occupies_at_time)
    summary.append({"check": "TotalEventsVsRooms", "category": "hard_explicit", "ok": ok, **meta})
    explicit_details["TotalEventsVsRooms"] = det

    # Soft explicit
    if hasattr(m1, "SpreadMinSoft") and hasattr(m1, "SpreadMaxSoft"):
        ok, det, meta = check_spread_soft(m1, event_to_course, time_to_day, tol=tol)
        summary.append({"check": "SpreadEventsSoft", "category": "soft_explicit", "ok": ok, **meta})
        explicit_details["SpreadEventsSoft"] = det

    group_specs = [
        # set_name, busy_name, present_name, under_name, over_name, min, max, penalty_name, label
        ("C1", "c1_busy", "c1_present", "c1_under", "c1_over", 3, 6, "PenaltyClass1Busy", "MinMaxADay_Classes_1"),
        ("C2", "c2_busy", "c2_present", "c2_under", "c2_over", 1, 2, "PenaltyClass2Busy", "MinMaxADay_Classes_2"),
        ("C3", "c3_busy", "c3_present", "c3_under", "c3_over", 1, 6, "PenaltyClass3Busy", "MinMaxADay_Classes_3"),
        ("C4", "c4_busy", "c4_present", "c4_under", "c4_over", 1, 7, "PenaltyClass4Busy", "MinMaxADay_Classes_4"),
        ("C5", "c5_busy", "c5_present", "c5_under", "c5_over", 3, 7, "PenaltyClass5Busy", "MinMaxADay_Classes_5"),
        ("T1", "t1_busy", "t1_present", "t1_under", "t1_over", 2, 7, "PenaltyTeacher1Busy", "MinMaxADay_Teachers_1"),
    ]
    for spec in group_specs:
        set_name, busy_name, present_name, under_name, over_name, mn, mx, pen_name, label = spec
        if hasattr(m1, set_name) and hasattr(m1, pen_name):
            ok, det, meta = check_minmax_group_soft(
                m1, set_name, busy_name, present_name, under_name, over_name,
                mn, mx, pen_name, tol=tol
            )
            summary.append({"check": label, "category": "soft_explicit", "ok": ok, **meta})
            explicit_details[label] = det

    # Implicit
    ok, det, meta = check_implicit_feasible_starts(E, feasible_starts)
    summary.append({"check": "Implicit_FeasibleStartsNonEmpty", "category": "implicit", "ok": ok, **meta})
    implicit_details["Implicit_FeasibleStartsNonEmpty"] = det

    ok, det, meta = check_implicit_unavailability(feasible_starts, occ_times, event_to_teachers, event_to_classes, resource_unavailability)
    summary.append({"check": "Implicit_UnavailabilityFiltered", "category": "implicit", "ok": ok, **meta})
    implicit_details["Implicit_UnavailabilityFiltered"] = det

    ok, det, meta = check_implicit_duration_day(feasible_starts, occ_times, event_to_duration, time_to_day)
    summary.append({"check": "Implicit_DurationAndDayBoundary", "category": "implicit", "ok": ok, **meta})
    implicit_details["Implicit_DurationAndDayBoundary"] = det

    ok, det, meta = check_implicit_feasible_rooms(E, feasible_rooms)
    summary.append({"check": "Implicit_FeasibleRoomsNonEmpty", "category": "implicit", "ok": ok, **meta})
    implicit_details["Implicit_FeasibleRoomsNonEmpty"] = det

    summary_df = pd.DataFrame(summary)
    if not summary_df.empty:
        summary_df = summary_df[["check", "category", "ok", "n_violations"] + [c for c in summary_df.columns if c not in ["check", "category", "ok", "n_violations"]]]

    return summary_df, explicit_details, implicit_details


# Example
summary_df, explicit_details, implicit_details = run_all_sanity_checks_elementary(
    m1=m1,
    E=E,
    T=T,
    R=R,
    feasible_starts=feasible_starts,
    occ_times=occ_times,
    occupies_at_time=occupies_at_time,
    time_to_day=time_to_day,
    event_to_duration=event_to_duration,
    event_to_teachers=event_to_teachers,
    event_to_classes=event_to_classes,
    event_to_course=event_to_course,
    resource_unavailability=resource_unavailability,  # dict: resource -> set(times)
    feasible_rooms=feasible_rooms if "feasible_rooms" in globals() else None,
    tol=1e-6
)

display(summary_df.sort_values(["category", "ok", "n_violations"], ascending=[True, True, False]))

# Optional: inspect details
for k, df in explicit_details.items():
    if len(df):
        print(f"\n{k} violations ({len(df)}):")
        display(df.head(20))

for k, df in implicit_details.items():
    if len(df):
        print(f"\n{k} violations ({len(df)}):")
        display(df.head(20))
```
